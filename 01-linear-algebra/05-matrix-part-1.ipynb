Запрограммируйте умножение матриц с использованием циклов for. Подтвердите свои результаты с помощью оператора @ библиотеки NumPy. Это упражнение поможет вам укрепить ваше понимание умножения матриц, но на практике всегда лучше использовать @ вместо написания двойного цикла for.

````python
A = np.arange(1, 10).reshape(3, 3)
B = np.arange(11, 20).reshape(3, 3)
print(A, B, sep="\n\n")
````

````python
def multiply_matrix(matrix_a, matrix_b):
    # Проверить, что М-строк matrix_1 == K-столбцов matrix_2
    a_row, a_col = matrix_a.shape[0], matrix_a.shape[1]
    b_row, b_col = matrix_b.shape[0], matrix_b.shape[1]

    if a_row != b_col:
        print(f"Нельзя умножить. Размерности матриц не совпадают. {matrix_a.shape} и {matrix_b.shape}")

    # Заготовка для результата
    result = np.zeros_like(matrix_a)

    # Перебираем строки/столбцы
    for i in range(a_row):
        row = matrix_a[i:i + 1, :][0]
        # print(row)

        for j in range(b_col):
            col = matrix_b[:, j]
            # print(col)

            value = np.sum([row[_] * col[_] for _ in range(len(row))])
            # result.append(value)

            result[i][j] = value

    return result
````

````python
print(multiply_matrix(A, B))
````

````python
print(A @ B)
````

Чтобы понять «магию» диагональных матриц, нужно на секунду забыть о громоздких формулах и посмотреть, как работает обычное умножение матрицы на вектор.Диагональная матрица — это по сути набор индивидуальных коэффициентов. Она не перемешивает данные (как обычные матрицы), а просто говорит каждому элементу: «Стань во столько-то раз больше».
1. Почему «слева» — это строки? ($D \cdot A$)Вспомни, как работает умножение матриц «строка на столбец». Когда диагональная матрица $D$ стоит слева:Чтобы получить первую строку результата, мы берем первое число из $D$ и умножаем его на всю первую строку матрицы $A$.Чтобы получить вторую строку — берем второе число из $D$ и умножаем его на всю вторую строку $A$.Простая интуиция: Матрица слева — это «фильтр» на входе. Она перехватывает каждую строку $A$ целиком и масштабирует её своим коэффициентом.
2. Почему «справа» — это столбцы? ($A \cdot D$)Теперь поменяем их местами. Когда $D$ стоит справа:По правилу умножения, первый столбец результата — это линейная комбинация столбцов $A$. Но так как в $D$ в первом столбце есть только одно число на первой позиции, оно «вытягивает» из $A$ только первый столбец и умножает его на себя.Второй элемент на диагонали $D$ «занимается» только вторым столбцом $A$.Простая интуиция: Матрица справа — это «настройка» выходов. Она берет готовые столбцы $A$ и подправляет их масштаб.
3. Магия «Сэндвича» ($D \cdot A \cdot D$)Когда мы делаем $D \cdot A \cdot D$, происходит двойная фильтрация. Представь таблицу умножения.Элемент на пересечении $i$-й строки и $j$-го столбца в матрице $A$ (допустим, там была единица):Сначала умножается на $i$-й элемент левой матрицы (масштаб строки).А потом умножается на $j$-й элемент правой матрицы (масштаб столбца).В итоге каждый элемент $a_{i,j}$ превращается в $a_{i,j} \cdot d_i \cdot d_j$.
