### Геометрия умножения вектора на скаляр

```python
# Создаем векторы
v = np.array([3, 2])
scalars = [1.0, 2.0, 0.33, 0.0, -0.67]
colors = sns.color_palette("husl", 5)

# Рисуем
fig, axes = plt.subplots(1, 5, figsize=(15, 3))

for i, (s, color) in enumerate(zip(scalars, colors)):
    ax = axes[i]
    v_scaled = v * s

    # Вектор
    ax.arrow(0, 0, v_scaled[0], v_scaled[1],
             head_width=0.3, head_length=0.4,
             fc=color, ec=color, linewidth=3)

    # Исходный вектор (пунктир)
    ax.arrow(0, 0, v[0], v[1],
             head_width=0.3, head_length=0.4,
             fc='gray', ec='gray', linestyle='--', alpha=0.5)

    # Настройки
    ax.set_xlim(-5, 7)
    ax.set_ylim(-4, 5)
    ax.axhline(0, color='k', alpha=0.2)
    ax.axvline(0, color='k', alpha=0.2)
    ax.set_aspect('equal')
    ax.set_title(f'×{s}', fontweight='bold')
    ax.grid(True, alpha=0.3)

# plt.suptitle('Умножение вектора на скаляр', fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

### 2.2
- Напишите алгоритм, который вычисляет норму вектора по формуле евклидова расстояния: квадратный
корень из суммы квадратов элементов вектора: $||\text{ v }||= \sqrt{\sum^{n}_{i=1}\text{v}^{2}_{i}}$

- Подтвердите, используя случайные векторы с разными размерностями и ориентациями, что вы получаете тот же результат, что и `np.linalg.norm()`.



```python
def get_norm(v):
    return np.sqrt(sum(x**2 for x in v))

for _ in range(11):
    # Генерируем случайный вектор
    v = np.random.randint(0, 101, size=np.random.randint(1, 11))
    my_norm = get_norm(v)
    np_norm = np.linalg.norm(v)
    print(f"{my_norm == np_norm}")
```

- Что происходит при вводе вектора нулей?

```python
v = np.random.randint(0, 101, size=np.random.randint(1, 11))
v_original_norm = np.linalg.norm(v)
v_norm = v / v_original_norm
v_norm_norm = np.linalg.norm(v_norm)

print(
    f"Исходный вектор {v}\n"
    f"Норма исходного вектора: {v_original_norm}\n"
    f"Единичный вектор {v_norm}\n"
    f"Норма единичного вектора: {v_norm_norm}"
)
```

```python
v_zero = np.array([0, 0, 0, 0])
norm_zero = np.linalg.norm(v_zero)

# С предупреждением о делении на ноль
with warnings.catch_warnings(record=True) as w:
    warnings.simplefilter("always")
    v_zero_unit = v_zero / norm_zero
    if w:
        print(f"Предупреждение: {w[0].message}")

print(f"Результат деления 0/0 в NumPy: {v_zero_unit}")
print(f"Элементы: {v_zero_unit}")
print(f"Содержит NaN? {np.any(np.isnan(v_zero_unit))}")
print(f"Содержит inf? {np.any(np.isinf(v_zero_unit))}")
```

### 2.4
Вы знаете, как создавать единичные векторы; что, если вы хотите создавать вектор любого произвольного модуля? Напишите функцию Python, которая на входе будет принимать вектор и желаемый модуль вектора и возвращать вектор в том же направлении, но с модулем, соответствующим второму входному аргументу.

**Разбираемся в деталях**
- Модуль (норма) вектора - это расстояние от хвоста до головы, которое считается по формуле евклидова расстояния.
- Единичный вектор - это вектор, модуль которого (норма) равен 1.
- Любой не единичный вектор имеет ассоциированный с ним единичный вектор.

**Надо сделать**
- Взять исходный вектор
- Взять желаемый модуль нового вектора
- Вернуть новый вектор в том же направлении, что и исходный, но с модулем равным желаемому.

**Как сделать**
- Исходный вектор привести к единичному
- Единичный вектор умножить на желаемый модуль

```python
def get_new_vector(base_vector, target_module):
    base_module = np.linalg.norm(base_vector)
    base_one = base_vector / base_module

    target_vector = base_one * target_module
    print(
        f"Base vector: {base_vector}\n"
        f"Base vector module: {base_module}\n"
        f"Base vector one: {base_one}\n"
        f"Target vector: {target_vector}\n"
        f"Target vector module: {np.linalg.norm(target_vector)}\n"
    )
```

```python
get_new_vector(base_vector=np.array([1,2,3]), target_module=5)
```

Напишите исходный код, чтобы продемонстрировать, что точечное произведение является коммутативным. Коммутативное свойство означает a×b=b×a, что для точечного произведения векторов означает aTb = bTa.

Продемонстрировав это в исходном коде, примените уравнение 2.9, чтобы понять, почему точечное произведение является коммутативным.

```python
# Комутативность
a = np.random.randn(5)
b = np.random.randn(5)

print(
    f"a: {a}\n"
    f"b: {b}\n"
    f"a*b=b*a: {np.dot(a, b) == np.dot(b, a)}"
)
```

```python
# Доказываем через точечное произведение
result = {
    "a*b":[],
    "b*a":[]
}

for ai, bi in zip(a, b):
    result["a*b"].append(ai * bi)
    result["b*a"].append(bi * ai)

sum(result["a*b"]) == sum(result["b*a"])
```

### 2.9
Реализуйте ортогональное разложение векторов. Начните с двух векторов случайных чисел t и r. Затем подтвердите, что сумма двух компонент равна t и что $t_{\bot r}$ и $t_{\parallel r}$ ортогональны.

**Разбираемся в деталях**
- Ортогональное разложение целевого вектора - это разбиение вектора на два других вектора таким образом
    - эти два вектора в сумме составляли целевой вектор
    - один вектор был ортогонален опорному вектору, а другой параллелен.
    - сумма новых векторов равна целевому вектору
- Вычисление параллельной компоненты
    - $\beta=\frac{t^{T}r}{r^{T}r}$
    - $t_{\parallel r} = r * \beta $
- Вычисление перпендикулярной компоненты $t_{\bot r} = t - t_{\parallel r}$
- Доказать ортогональность $(t_{\bot r})^{T}r=0$

```python
t = np.random.randn(2)
r = np.random.randn(2)

# Находим скаляр (бета)
# np.dot(t, r) - это и есть t^T * r в математической записи
# np.dot(r, r) - это и есть r^T * r в математической заиси
beta = np.dot(t, r) / np.dot(r, r)

# Параллельная компонента: вектор r, умноженный на число beta
t_par = beta * r

# Перпендикулярная
t_per = t - t_par

print(
    f"Сумма двух компонент равна исходному вектору: {sum(t) == sum(t_par) + sum(t_per)}\n"
    f"Ортогональность: {np.isclose(t_par @ t_per, 0)} "
)
```
